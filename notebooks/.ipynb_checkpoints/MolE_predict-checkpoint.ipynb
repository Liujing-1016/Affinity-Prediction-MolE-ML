```python
def filter_smile_row(df0, smile, threshold=0.1):
    """
    Filter the cells andnn corresponding target id that are greater than the threshold in a given SMILE row. 
    
    Parameters：
        df0(pd.read_csv):  
        smile (str): must be in the dataframe. 
        threshold (float): pIC50 values, 
        
    """

    # find SMILE line
    row = df0[df0['smiles'] == smile]
    if row.empty:
        print(f"Can't find the SMILE: {smile}")
        return {}

    # Filter pIC50 values greater than threshold, excluding 'smiles' column
    row_values = row.iloc[0, 1:]  # Get values from second column onward
    filtered = row_values[row_values > threshold]

    result = filtered.to_dict()
    if result:
        print(f"When pIC50 values are greater than {threshold}:")
        for key, value in result.items():
            print(f"Target {key}: {value}")
    else:
        print(f"No pIC50 values greater than {threshold} found for SMILE: {smile}")

    return result
```

```python
import pandas as pd
import numpy as np

def filter_by_target_index(df0, target_idx, threshold=0.1):

    target_columns = df0.columns[1:]

    if target_idx < 0 or target_idx >= len(target_columns):
        print(f"Invalid target index: {target_idx}. It must be between 0 and {len(target_columns)-1}.")
        return pd.DataFrame()
    
    selected_target = target_columns[target_idx]
    print(f"\nSelected target: {selected_target} (index {target_idx})")
    
    # Filter rows where pIC50 > threshold, handle NaN values
    try:
        filtered = df0[df0[selected_target].notna() & (df0[selected_target] > threshold)][['smiles', selected_target]]
    except (TypeError, ValueError) as e:
        print(f"Error filtering data: {e}. Ensure {selected_target} contains numeric values.")
        return pd.DataFrame()

    if not filtered.empty:
        # Add target_idx as a column for clarity
        filtered['target_idx'] = target_idx
        print(f"Molecules with pIC50 > {threshold} for target {selected_target} (index {target_idx}):")
        print(filtered)
    else:
        print(f"No molecules found with pIC50 > {threshold} for target {selected_target} (index {target_idx}).")
    
    return filtered
```

```python
def load_target_dict(file_path):
    """
    txt dictionary
    """
    with open(file_path, 'r') as file:
        content = file.read()
        target_dict = eval(content)  
    
    return target_dict

def get_target_name_by_index(target_dict, target_idx):
    """
    get target name by index
    """
    target_names = list(target_dict.keys())
    
    if 0 <= target_idx < len(target_names):
        return target_names[target_idx]
    else:
        return f"Invalid index: {target_idx}. Valid range is 0 to {len(target_names)-1}."

dict_path = "/home/jovyan/proj-liujing/Data_preparation/target_to_index_dictionary_unsorted.txt"
target_dict = load_target_dict(dict_path)
```
